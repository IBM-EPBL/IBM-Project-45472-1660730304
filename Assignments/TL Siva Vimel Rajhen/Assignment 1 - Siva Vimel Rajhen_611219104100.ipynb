## 5. Create an array of all the even integers from 20 to 35

```python
#Array with even integers from 20-35
import numpy as np
n = np.arange(20,35,2) #'2' for even splitting between numbers from '20-35'
n
```

## 6. Create a 3x3 matrix with values ranging from 0 to 8

```python
#3x3 matrix ranging from 0-8
import numpy as np
n = np.arange(0,9).reshape(3,3) #arrage(0,9)- for values ranging from 0-8 & reshape(3,3)- for matrix shape as 3x3 
n
```

## 7. Concatenate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
#Answer for a & b concatenation
import numpy as np
a = np.array([1,2,3]) #a
b = np.array([4,5,6]) #b
n = np.concatenate((a,b),axis=None) #axis=None for printing list in series
n
```

## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
#Dates from 1-January-2023 to 10-February-2023
import pandas as pd
date = pd.date_range(start='01-01-2023',end='10-02-2023')
for val in date: #to iterate dates ranging from jan to feb
  print(val)
```
